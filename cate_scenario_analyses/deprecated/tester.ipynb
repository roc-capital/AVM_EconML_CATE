```python
'''PROPERTYID – Internal ID that uniquely identifies the property in the dataset.

HPI_MID – House Price Index value for the property’s area (a market-level price index used to trend/adjust values).

FIPS – 5-digit federal county code that identifies the county the property is in.

ZIP_COAL – Standardized 5-digit ZIP code for the property, merged from multiple sources.

SALEAMT – Recorded sale price for the most recent transaction (in dollars).

YEAR – Calendar year for this record (e.g., year of observation/modeling).

LATITUDE – Geographic latitude of the property.

LONGITUDE – Geographic longitude of the property.

BEDROOMS_MLS – Number of bedrooms from the MLS listing.

ROOMS_COAL – Best available total number of rooms in the home, combined across sources.

NUM_PRIOR_MLS_LISTINGS – How many times this property has appeared as a prior MLS listing.

PREV_SALEYEAR_COAL – Year of the prior recorded sale, using the best available data.

TRENDED_PRIOR_SALE_MLS – Prior MLS sale price adjusted (“trended”) to current dollars using a price index.

MOST_RECENT_SALE_COAL – Best estimate of the most recent sale price, merged across sources.

FULL_BATHS_COAL – Number of full bathrooms (with a tub or shower) in the home.

HALF_BATHS_COAL – Number of half bathrooms (toilet + sink only).

LIVINGAREASQFT_COAL – Best available square footage of the main living area.

LOTSIZESQFT_COAL – Best available lot size in square feet.

YEARBUILT_COAL – Best estimate of the original year the home was built.

EFFECTIVEYEARBUILT_COAL – “Effective” build year reflecting major renovations/updates.

BUILDINGAREA_ASSESSOR – Building square footage from the assessor (assessment record).

GARAGE_SPACES_COAL – Number of parking spaces in the garage.

FIREPLACE_TYPE_MLS – Fireplace type from MLS (e.g., gas, wood-burning).

FIREPLACE_COUNT_MLS – Number of fireplaces from MLS data.

PORCH_OR_PATIO_MLS – Indicator/details from MLS that the home has a porch and/or patio.

LAND_USE_ASSESSOR – Assessor’s land use code (e.g., single-family, condo, commercial).

SECTION – Legal land-description “section” code indicating which section of the city/subdivision the property is in.

TOTAL_BATHS_COAL – Combined total bathroom count (full + partial, often in a numeric like 2.5).'''
```
